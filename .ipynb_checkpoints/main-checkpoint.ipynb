## Making the Convolution Biologically Plausible for SNN Image Recognition

We typically implement convolutions as a multiplication between a kernel and the subsets of the image it can project to.

In this notebook we try to show that the non-memoryless-ness of the LIF model can be leveraged to represent a pseudo-convolution operation, done entirely in the additive spiking domain, and without calling upon any (biologically implausible) multiplication.


We think this could be especially useful to translate the idea of convolutional layers (which perform better for feature detection and hence image recognition) into biologically sound neural networks models.

### __In this case, the left layer represents different sensory neurons and their 'weights' (spike interval), and the kernel in the middle represents any _additional_ effect caused by non-sensory feedback neurons/dendrites to their respective sensory neurons.__

Unlike typical convolutional neural networks, we don't have a kernel which just "travels" across the image. Instead, there are many stationary kernel across the same image. By stimulating the right 'kernel' dendrites/feedback neurons in interesting ways, we are able to extract features from the image.

### __We seek to define a 'convolution' operation in the spiking domain -with an output is the later layer neurons-, without using the traditional dot product method shown in the picture, but only with addition in leaky neuron models__

## A Review of the Standard Convolution: Edge Detection

```python
import torch

test_im =  torch.tensor([[7, 7, 7, 0, 0, 0],
                         [7, 7, 7, 0, 0, 0],
                         [7, 7, 7, 0, 0, 0],
                         [7, 7, 7, 0, 0, 0],
                         [7, 7, 7, 0, 0, 0],
                         [7, 7, 7, 0, 0, 0]]) #  image with an edge in the middle

control_im=torch.tensor([[7, 7, 7, 7, 7, 7],
                         [7, 7, 7, 7, 7, 7],
                         [7, 7, 7, 7, 7, 7],
                         [7, 7, 7, 7, 7, 7],
                         [7, 7, 7, 7, 7, 7],
                         [7, 7, 7, 7, 7, 7]]) #  image with no edge

size_im = test_im.size()[0] #  equal to 6 b/c 6x6 square image


#  right-going edge detector kernel
test_kernel = torch.tensor([[1, 0, -1],
                            [1, 0, -1],
                            [1, 0, -1]])

size_ker = test_kernel.size()[0] #  equal to 3 b/c 3x3 square kernel

#--------------------------------#

size_con = size_im-size_ker+1 #Size of the convolution output (if stride=1)
convolved1 = torch.zeros((size_con, size_con))
convolved2 = convolved1.clone()

#  applying convolution to 1st image
for iii in range(size_con):
    for jjj in range(size_con):
        elementwise_mult = test_kernel*test_im[iii:iii+size_ker, jjj:jjj+size_ker]
        convolved1[iii, jjj] = elementwise_mult.sum()
        
#  same to second image
for iii in range(size_con):
    for jjj in range(size_con):
        elementwise_mult = test_kernel*control_im[iii:iii+size_ker, jjj:jjj+size_ker]
        convolved2[iii, jjj] = elementwise_mult.sum()

print(convolved1.numpy(), end=' ')
print("--> 21 is large, i.e high-contrast edge detected in the middle of the image\n")
print(convolved2.numpy(), end=' ')
print("--> no edge detected")
```

## Frequency-sum based Convolution on LIF

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

#LIF model copied from neuromatchacademy. Uses Euler's method
#https://compneuro.neuromatch.io/tutorials/W2D3_BiologicalNeuronModels/studentW2D3_Tutorial1.html
import lif_model
```

```python
physical_params = lif_model.default_pars(T=150)
```

```python
def add_to_class(Class):
    """Register functions as methods in created class."""
    def wrapper(obj):
        setattr(Class, obj.__name__, obj)
    return wrapper
```

### Create an array 6 neurons. 100Hz means light, 50Hz means dark

```python
from scipy.special import softmax

class N_Container():
    class N():
        def __init__(self, input_current, params='default'):
            if params == 'default':
                self.physical_params = lif_model.default_pars(T=150)
            else:
                self_physical_params = params

            self.input_current = input_current
            self.voltage, self.output_spikes = lif_model.run_LIF(physical_params, Iinj=input_current, stop=False)
            self.avg_out = len(self.output_spikes)/150

        def plot(self):
            lif_model.plot_volt_trace(self.physical_params, self.voltage, self.output_spikes)

        def __float__(self):
            return self.avg_out
        
    def __init__(self, size, frequency_array):
        f = {0: 0, 50: 300, 100 : 400} #frequency dictionary- turns average spikes into current
        assert isinstance(size, tuple)
        assert isinstance(frequency_array, np.ndarray)
        
        dim = size[0], size[1]
        assert (dim == frequency_array.shape)
        self.arr = np.array([N(f[freq]) for freq in frequency_array.flatten()]).reshape(dim)
    
    def __iter__(self):
        return self.arr
    
    def __getitem__(self, index):
        return self.arr[index]
    
    def get(self):
        output = np.zeros_like(self.arr)
        for iii, line in enumerate(self.arr):
            for jjj, neuron in enumerate(line):
                output[iii][jjj] = float(neuron)
        arr = np.floor(output*400)
        return np.array(arr, dtype=np.float64)
    
    def getsoftmax(self):
        print(np.floor(255*softmax(self.get())))
        return np.floor(255*softmax(self.get()))
    
    def __len__(self):
        return len(self.arr)
```

```python
freq_topology = np.array(([100, 100, 100, 50, 50, 50]*6)).reshape((6, 6)) # times 6 rows. dont forget double bracks
dummy_topology= np.array(([100, 100, 100, 100, 100, 100]*6)).reshape((6, 6))

neurons1 = N_Container((6, 6), freq_topology)
neurons2 = N_Container((6, 6), dummy_topology)
plt.subplot(1, 2, 1); plt.imshow(neurons1.get(),vmin=0, vmax=100); plt.subplot(1, 2, 2); plt.imshow(neurons2.get(), vmin=0, vmax=100)
```

```python
@add_to_class(N)
def __add__(self, other_neuron):
    
    assert isinstance(other_neuron, N)
    print(1)
    #addition outputs dummy neuron with input as the sum of currents
    return N(self.input_current+other_neuron.input_current )
```

### We first try with the 0 kernel ("identity" kernel)
The output layer is formed by adding the inputs of its projecting neurons

```python
ker = np.array(([0]*9)).reshape((3, 3))
kernel = N_Container((3, 3), ker)

size_output = len(neurons1)-len(kernel)+1
out = np.array(([0]*(size_output**2))).reshape((size_output, size_output))
output = N_Container((size_output, size_output), out)

#step through iteration: for each output neuron, sum 
for iii in range(size_output-1):
    for jjj in range(size_output-1):
        for aaa in kernel[iii]
            for bbb in aaa:
                output[iii][jjj] = output[iii][jjj]+ neurons1[iii][jjj] + kernel[iii][jjj]

plt.imshow(neurons1.get())
plt.subplot(1, 2, 1); plt.imshow(output.getsoftmax());
```
